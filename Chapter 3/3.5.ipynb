Consider the SCM
$$\begin{align*}X&:=Y^2+N_X\\ Y&:=N_Y\end{align*}$$
with $N_X, N_Y\sim \mathcal{N}(0,1)$. Generate an i.i.d samples of size 200 from the joint distribution $(X,Y)$

```python
import numpy as np
import pandas as pd
```

```python
y = np.random.randn(200)
x = y*y+np.random.randn(200)

samples = pd.DataFrame({"X": x, "Y": y})
```

```python
samples.plot(kind='scatter', x='Y',y='X')
```
