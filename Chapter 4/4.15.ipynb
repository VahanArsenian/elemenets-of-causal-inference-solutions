```python
import numpy as np
import seaborn as sb
```

a)

```python
sp_size = 10_000
x =  np.random.uniform(1, 3, sp_size)
y = 2*x+np.random.uniform(-0.5, 0.5, sp_size)
```

```python
sb.kdeplot(x=x, y=y, fill=True, thresh=0, levels=100, cmap="mako")
```

From the KDPlot we can easily see the support of $P(X, Y)$. Assume there exists an ANM from $Y$ to $X$. When $M_Y=6.5\implies Y=6.5\implies X=3\implies M_X=0$ with probability 1. However, for $M_Y=2\implies Y=2$ $M_X$ does not bound to $0$ and can vary within a range. Thus, different values $M_Y$ yield different support for $M_X$ violating the independence assumption of $M_X$ and $M_Y$.
